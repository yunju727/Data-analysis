## Data Dictionary
* Survived: 0 = No, 1 = Yes
* pclass : Ticket class 1 = 1st, 2 = 2nd, 3 = 3nd
* sibsp : #of siblings / spouses aboard the Titanic     #같이 탄 승객 수
* parch : #of parents / children aboard the Titanic     
* ticket : Ticket number
* cabin : Cabin number
* embarked : Port of Embarkation C = Cherbourg, Q = Queenstown, S = Southampton


### 결측치? NaN ?
* 데이터 셋을 다루다보면 해당하는 열의 값을 갖지 않는 경우가 있는데, 아무런 값을 갖지 않는 것을 결측치라고 한다.
* 판다스에서는 이러한 결측치를 NaN으로 표시한다. 
* 결측치는 예측가능한 경우에는 예측값을 넣어주거나 예측불가능한 경우에는 제거하는 등 처리를 해주어야 나중에 분석시 문제가 발생하지 않는다.
* 예측가능한 경우? 1 2 3 4 NaN => ?
* 예측불가능한 경우 ? 위의 Cabin같은 데이터 (이름, 티켓정보만 보고 캐비넷 번호를 예측할 수 없기 때문에)
